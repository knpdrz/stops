```python
import numpy as np
import csv
import pandas as pd
import nltk
# https://www.nltk.org/book/ch01.html
```

```python
with open('allStops.csv', newline='', encoding='UTF-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    linesNum = sum(1 for row in spamreader)
    print(linesNum)
    for row in spamreader:
        print(', '.join(row))
```

```python
df = pd.read_csv('allStops.csv')
#print(df["Stop name"])

fdist = nltk.FreqDist(df["Stop name"])
fdist.most_common
```

```python
fdist.plot(15, cumulative=False, title="Częstość występowania nazw ")
```

```python
firstStopLetters = [stopName[0] for stopName in df["Stop name"]]
fdistFirstLetter = nltk.FreqDist(firstStopLetters)
fdistFirstLetter.plot(20, title="Częstość występowania pierwszych liter w nazwie przystanku")
```
